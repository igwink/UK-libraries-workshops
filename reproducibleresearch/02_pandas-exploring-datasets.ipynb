## A Workflow for Exploring Data

Whenever you encounter a new dataset, you can follow the same basic sequence of steps, and this notebook is organized around them:

1. **Get data into pandas** (build it by hand, or load it from a file or the web)
2. **Inspect its structure** (How big is it? What are the columns?)
3. **Ask questions of it** (What values occur? How often? What's the largest/smallest?)
4. **Save what you found**

We'll learn each step using a toy dataset, and then at the end of the session you'll apply the whole workflow to a real, live dataset: every earthquake detected on Earth in the past week.

## The Pandas Series and DataFrame
In Pandas, data are stored in two fundamental objects:

* **Pandas Series** - a single column or row of data
* **Pandas DataFrame** - a table of data containing multiple columns and rows

### Pandas Series

We can think of a Series as a single column or row of data. Here we have a column called `Champions` with the country names of the winners of the most recent ten FIFA world cup games.

|Champions|
|---|
|Argentina|
|France|
|Germany|

|Italy|
|Brazil|

|Brazil|

Let's create a Series based on this column. To create our Series, we pass a **list** into the `.Series()` method:

```python
# Create a data series object in Pandas
champions = pd.Series(["Argentina",
                       "France",
                       "Germany",
                       "Spain",
                       "Italy",
                       "Brazil",
                       "France",
                       "Brazil",
                       "Germany",
                       "Argentina"]
                    )
```

```python
champions
```

In some ways, Series objects behave similar to lists in Python. For example, you can access a particular value in a Series using its index inside square brackets, or you can iterate over it using a `for` loop, just like we did in Session 1:

```python
# Access an item in the series using its index:
champions[4]
```

```python
for winner in champions:
    print(winner)
```

### Pandas DataFrame

While a Pandas Series is 1-dimensional with a single column/row of data, a Pandas DataFrame is 2-dimensional and can have multiple columns and rows.

|Year|Champion|Host|

Let's create a Pandas DataFrame based on this table. To create our dataframe, we pass lists of three items each into the `DataFrame()` method. Each of these lists is one row of the DataFrame. We also set a list of `columns` to specify the name for each column.

```python
# Create a Pandas dataframe
wcup = pd.DataFrame(
    data=[
        [2022, "Argentina", "Qatar"],
        [2018, "France", "Russia"],
        [2014, "Germany", "Brazil"],
        [2010, "Spain", "South Africa"],
        [2006, "Italy", "Germany"],
        [2002, "Brazil", "Korea/Japan"],
        [1998, "France", "France"],
        [1994, "Brazil", "USA"],
        [1990, "Germany", "Italy"],
        [1986, "Argentina", "Mexico"]
    ],
    columns=["Year", "Winner", "Host"]
)

wcup
```

### Sorting a DataFrame

To reorder the rows in a DataFrame, we can use `.sort_values(by=column)`, specifying the name of the column we would like to use to sort the data. Sorting answers questions like, "What was the earliest World Cup in our data?"

In order to sort by a column, all the data in that column must be of the same (or comparable) type. For example, strings can all be sorted alphabetically, or numbers (floats and integers) can be sorted numerically.

```python
# Sort the DataFrame by year:
wcup.sort_values(by="Year")
```

`.sort_values()` has an optional additional parameter, `ascending`. By default, pandas sorts a DataFrame in ascending order (e.g. from A to Z or from lowest number to highest number). To reverse the order, we can include `ascending=False` in the function call.

```python
# Sort the DataFrame in reverse alphabetical order by winner:
wcup.sort_values(by="Winner", ascending=False)
```

### Subsetting a DataFrame Based on a Condition

Oftentimes, we will want to work with only a subset of our DataFrame that meets some condition. For example, we may want to only work with rows in our `wcup` DataFrame where Germany was the winner, or only games played after 2000.

We do this in two steps. First, we write a comparison operator expression to check which rows in a column meet a given condition.

For example, the line below checks for rows where the value of "Winner" is equal to "Germany". In Session 1, a comparison like `my_name == your_name` returned a single `True` or `False`. Here, because we're comparing a whole *column*, we get back a whole Series of boolean values — `True` for rows where the condition is met and `False` for rows where it is not:

```python
wcup["Winner"] == "Germany"
```

This Series of True/False values is sometimes called a **mask**. When we pass it into our original DataFrame, it will return only the rows where the condition is met.

```python
# Step 1: Create a mask for the condition we want to use to subset our data:
mask = wcup["Winner"] == "Germany"

# Step 2: Pass the mask to the original DataFrame:
wcup[mask]
```

## Try It Out: Exploring Live Earthquake Data

So far we've worked with a small table we typed in by hand. Real datasets usually come from somewhere else, such as a file a colleague sent you, a database, or a live source on the web.

The US Geological Survey (USGS) publishes a continuously updated feed of every earthquake detected worldwide. You can view the feeds and learn more about the structure of data they return [here](https://earthquake.usgs.gov/earthquakes/feed/v1.0/geojson.php).

You can take a look at the data in its raw form in your browser here: https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_week.geojson

We'll use the `requests` library to download the past week of earthquakes. Its `.json()` method converts the downloaded data into the Python data types you already know from Session 1: **dictionaries** and **lists**.

```python
# Import the requests library for accessing data from the Internet:
import requests

# Download the past 7 days of earthquakes from the USGS:
url = "https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_week.geojson"
data = requests.get(url).json()

# The result is a dictionary:
type(data)
```

```python
# The dictionary's "features" key holds a list with one item per earthquake.
# How many earthquakes were detected in the past week?
len(data["features"])
```

```python
# Each item in that list is itself a dictionary, and the information we
# want is stored under its "properties" key. Let's look at one earthquake:
data["features"][0]["properties"]
```

To turn this into a DataFrame, we need a list containing the properties of every earthquake. We can build one using a `for` loop and `.append()`:

```python
# Collect the properties of every earthquake into a list:
quake_list = []
for event in data["features"]:
    quake_list.append(event["properties"])

# Turn the list into a DataFrame:
quakes = pd.DataFrame(quake_list)

quakes.head()
```

# Using AI Tools for Data Exploration and Analysis

Coding agents and LLM-powered chatbots have made coding significantly more accessible, given that you can explain in natural language what you would like to happen in code. However, in a research context, code not only helps you produce a result but also records the steps you took to get there. As a result, researchers should make sure they can interpret any code they have an AI tool produce for them.

One specific area that's gotten easier for researchers is working with library documentation. While something like the [official pandas documentation](https://pandas.pydata.org/docs/) will always be an important resource, for most beginner questions the code produced by Google Gemini in response to a search can tell you what you need to know.

If you'd like to try out how AI can extend your understanding of pandas, try using your preferred tool to learn how to do the following and apply it to the `quakes` dataframe:
- Display the `value_counts()` for the `felt` column including missing values. (By default, `value_counts()` excludes missing `NaN` values.)
- Create a histogram of the values in the `mag` column.
- Split the `place` column into two columns, one containing the distance and the second containing the place name. (**Hint:** Notice that all the strings in the `place` column follow the same structure. Try identifying a piece of text you could use to separate the distance from the place name.)

Here are some general principles to consider when incorporating AI into your data workflow:
1. **Have the tool produce code rather than asking it to clean or analyze your data.** Researchers commonly ask AI tools to clean messy datasets for them, but this has two main weaknesses. First, the choices the AI made in cleaning the data often aren't visible, and second, cleaning future datasets requires future use of the AI, which may not be available in the same form in the future. By asking an AI tool to write code to clean your data instead, you solve both of these problems.
2. **Request code written in programming languages you're already familiar with.** A basic understanding of a programming language (like the one you gain in these workshops) can help you formulate better prompts by referring specifically to the libraries or data types you're working with, and it can help you catch errors more easily.
3. **Build your understanding by asking for code frameworks rather than the code for your specific circumstance.** Manually adapting an AI-written template to fit it into your workflow helps focus your attention on individual steps and understand how they fit together (rather than pasting a big block of code with no understanding of what it's doing).
