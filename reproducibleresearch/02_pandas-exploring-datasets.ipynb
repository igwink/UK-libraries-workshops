## Importing the Pandas Library

Before we can get started with pandas, we first need to import the library into our Python workspace. We do so with the following line of code:

```python
import pandas as pd
```

The code `import pandas as pd` instructs Python to do two things:
- Load the pandas library into our current workspace
- Assign pandas to `pd`. Now, whenever we reference `pd`, Python will know that we are accessing some capability of pandas.

## The Pandas Series and DataFrame
In Pandas, data are stored in two fundamental objects: 

* **Pandas Series** - a single column or row of data
* **Pandas DataFrame** - a table of data containing multiple columns and rows

### Pandas Series

We can think of a Series as a single column or row of data. Here we have a column called `Champions` with the country names of the winners of the most recent ten FIFA world cup games.

|Champions|
|---|
|Argentina|
|France|
|Germany|
|Spain|
|Italy|
|Brazil|
|France|
|Brazil|
|Germany|
|Argentina|

Let's create a Series based on this column. To create our Series, we pass a **list** into the `.Series()` method:

```python
# Create a data series object in Pandas
champions = pd.Series(["Argentina",
                       "France", 
                       "Germany", 
                       "Spain", 
                       "Italy", 
                       "Brazil", 
                       "France", 
                       "Brazil", 
                       "Germany",
                       "Argentina"]
                    )
```

```python
champions
```

In some ways, Series objects behave similar to lists in Python. For example, you can access a particular value in a Series using its index inside square brackets, or you can iterate over it using a `for` loop:

```python
# Access an item in the series using its index:
champions[4]
```

```python
for winner in champions:
    print(winner)
```

### Pandas DataFrame

While a Pandas Series is 1-dimensional with a single column/row of data, a Pandas DataFrame is 2-dimensional and can have multiple columns and rows. 

|Year|Champion|Host|
|---|---|---|
|2022|Argentina|Qatar|
|2018|France|Russia|
|2014|Germany|Brazil|
|2010|Spain|South Africa|
|2006|Italy|Germany|
|2002|Brazil|Korea/Japan|
|1998|France|France|
|1994|Brazil|USA|
|1990|Germany|Italy|
|1986|Argentina|Mexico|

Let's create a Pandas DataFrame based on this table. To create our dataframe, we pass lists of three items each into the `DataFrame()` method. Each of these lists is one row of the DataFrame. We also set a list of `columns` to specify the name for each column.

```python
# Create a Pandas dataframe
wcup = pd.DataFrame(
    data=[
        [2022, "Argentina", "Qatar"],
        [2018, "France", "Russia"],
        [2014, "Germany", "Brazil"],
        [2010, "Spain", "South Africa"],
        [2006, "Italy", "Germany"],
        [2002, "Brazil", "Korea/Japan"],
        [1998, "France", "France"],
        [1994, "Brazil", "USA"],
        [1990, "Germany", "Italy"],
        [1986, "Argentina", "Mexico"]
    ],
    columns=["Year", "Winner", "Host"]
)

wcup
```

### The Index

Note that in addition to the three columns we created, the DataFrame displayed a series of integers 0-9 on the lefthand side. This is the DataFrame's **index**. By default, the index is a series of integers starting with 0, but it is possible to set other values for our index (for example, we could use the Year column as our index). 

We will revisit the index when we look at accessing specific rows or columns of our data.

## Selecting a Specific "Cell"

If you've worked in Excel previously, you are likely familiar with referring to data points as "cells," where each cell can be referred to by a combination of its column (a letter) and its row (a number). For example, for a simple table in Excel, data might be laid out like this: 
||A|B|C|
|---|---|---|---|
|1|A1|B1|C1|
|2|A2|B2|C2|
|3|A3|B3|C3|

While pandas users don't typically refer to individual cells, it can be a helpful way of thinking about how data in a DataFrame is laid out. We can reference and select a specific cell by referring to it using its column name and index name with the following syntax:

`dataframe_name[column][index]`

In our `wcup` DataFrame, our column names are strings (text), and our index is composed of integers (numbers). Remember that in Python, indexes start at 0, so to select the first value in the "Winner" column, we would use the following code:

```python
# Select the value in the "Winner" column at index 0:
wcup["Winner"][0]
```

## `.value_counts()` and `.unique()`

We've looked at some ways to explore a DataFrame as a whole, but we can dig even more into individual columns. Once we have subset our DataFrame to a single column, we can use `.value_counts()` to get a table of the number of times that each value occurs in the column. We can also get a list of unique values (without their counts) using `.unique()`.

```python
wcup["Winner"].value_counts()
```

```python
wcup["Host"].unique()
```

## Sorting a DataFrame

To reorder the rows in a DataFrame, we can use `.sort_values(by=column)`, specifying the name of the column we would like to use to sort the data.

In order to sort by a column, all the data in that column must be of the same (or comparable) type. For example, strings can all be sorted alphabetically, or numbers (floats and integers) can be sorted numerically.

```python
# Sort the DataFrame by year:
wcup.sort_values(by="Year")
```

`.sort_values()` has an optional additional parameter, `ascending`. By default, pandas sorts a DataFrame in ascending order (e.g. from A to Z or from lowest number to highest number). To reverse the order, we can include `ascending=False` in the function call.

```python
# Sort the DataFrame in reverse alphabetical order by winner:
wcup.sort_values(by="Winner", ascending=False)
```

## Subsetting a DataFrame Based on a Condition

Oftentimes, we will want to work with only a subset of our DataFrame that meets some condition. For example, we may want to only work with rows in our `wcup` DataFrame where Germany was the winner, or only games played after 2000.

We do this in two steps. First, we write a **comparison operator** expression to check which rows in a column meet a given condition.

For example, the line below checks for rows where the value of "Winner" is equal to "Germany" and returns a Series containing the boolean values True (for rows where the condition is met) and False (for rows where the condition is not met).

```python
wcup["Winner"] == "Germany"
```

This Series of True/False values is sometimes called a **mask**. When we pass it into our original DataFrame, it will return only the rows where the condition is met.

```python
# Step 1: Create a mask for the condition we want to use to subset our data:
mask = wcup["Winner"] == "Germany"

# Step 2: Pass the mask to the original DataFrame:
wcup[mask]
```

## Writing a DataFrame to a File

Now that we've created our `wcup` DataFrame, let's save it to a file so that we can use it in other programs. One of the most common formats for saving tabular data is as a CSV file, because it is a simple format that can be used in a wide variety of software environments, including Excel.

To save a DataFrame to a file, we will use the method `.to_csv()`. 

Specifically, we will write `wcup.to_csv("wcup_winners.csv", index=False)`. The name for our file should always end in `.csv`.

```python
wcup.to_csv("wcup_winners.csv", index=False)
```

Why do we specify `index=False`? In this case, our index is just a default list of integers and isn't useful for working with the data, so we can drop it.
