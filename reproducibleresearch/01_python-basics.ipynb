## Variables

Oftentimes when working in Python, you will need to store some piece of data so that you can use it again later. You can do this by assigning that data to a variable.

We create (or *initialize*) a variable using the `=` sign. Once we have assigned a value to a variable, Python will use that value whenever the variable is called (unless we change it).

````python
# Create a variable called A, and assign an integer to it:
A = 20
````

Notice that unlike the `print("Hello World")`code we've run so far, the cell containing `A = 20` didn't produce any output. That's OK! Our variable has still been saved, and now we can use it:

````python
# Print the variable we just created:
print(A)
````

Python now treats `A` exactly as though it were the number 20, and we can use it to do anything we could do with a number:

````python
# Subtract 10 from our variable A:
A - 10
````

Even though we reused the variable `A` in a new line of code, we didn't change its value. To do that, we would need to assign a new value with a new `=` statement:

````python
# Print a sum that uses our variable A:
print(A + 10)

# Assign a new value to A:
A = 5

# Print a sum with A again:
print(A + 10)
````

##### Functions
Many different kinds of programs often need to do very similar operations. Instead of writing the same code over again, you can use a function. Essentially, a **function** is a small snippet of code that can be run using its name. There are three kinds of functions:

* Native functions built into Python
* Functions others have written that you can import
* Functions you write yourself

Most functions accept **arguments**, pieces of data that are used in the function's code. To use a function, you typically write its name, followed by parentheses with the arguments you want to use inside.

In most code documentation, you can recognize a function by the fact that is name is followed by parentheses, such as `len()`.

We've already been using one of the most common functions in Python, `print()`, which simply prints an object, such as a string.

````python
# Use the print() function on a string:
print("Hello, World!")
````

`print()` is one of the most versatile native functions in Python. Another useful one is `type()`, which lets us check what type of data an object is.

### Data Type 4: Lists

A **list** is an object that stores and orders other objects: integers, floats, strings, or even other lists.

A list assignment statement takes the form:
`my_list = [item1, item2, item3, item4...]`

````python
# A list containing integers
my_favorite_numbers = [13, 87, 89]
print(my_favorite_numbers)

# A list containing strings:
my_elements = ["Aluminum", "Francium", "Actinium"]
print(my_elements)
````

Both `my_favorite_numbers` and `my_elements` have three items, but we could have also created a list with no items. If we start with an empty list, we can add additional items to it using the `.append()` method. Whatever we put inside the parentheses will be added to the list:

````python
# Create an empty list:
reminders = []

# Add items to the list with .append(), printing the list in between additions:
reminders.append("Take out the trash")
print(reminders)
reminders.append("Empty the dishwasher")
print(reminders)
````

#### Some Methods Change an Object in Place; Others Hand You Back a New One

Compare `.lower()` and `.append()`. They look similar — both are methods called with `object.method()` — but they behave in opposite ways.

`.lower()` does **not** change the original string. Strings can't be modified once created, so `.lower()` leaves the original untouched and returns a brand-new string. If you want to keep the result, you have to catch it in a variable:

````python
school = "University of Kentucky"
school.lower()        # produces "university of kentucky", but doesn't save it
print(school)         # still "University of Kentucky" — unchanged!

school = school.lower()   # reassign to actually keep the change
print(school)             # now "university of kentucky"
````

`.append()` works the opposite way. It changes the list **in place**, meaning that the original list is modified directly. So you do **not** reassign it to a variable:

````python
reminders = []
reminders.append("Take out the trash")
print(reminders)
````

In **most** cases, functions return a new value and leave the original input unchanged. If you're ever unsure, however, you can test out the function in a cell or look up its documentation.

#### What You Can Do With Lists: `for` Loops

In the challenge above, you had to write a separate line of code for every item in the list. That gets old fast — imagine a list with 10,000 items! When we want to do something with **every item** in a list, we use a `for` loop, which repeats a block of code once for each item:

```
for item in list_name:
    do something with item
```

Two details to notice:
- `item` is a new variable name that **we choose**. One at a time, each value in the list gets assigned to it.
- The indented line (or lines) after the colon are the **code block** that gets repeated. Python uses indentation to know which code belongs inside the loop.

````python
# Print every topping in pizza_toppings:
for topping in pizza_toppings:
    print(topping)
````

````python
# Loops can do more than print. Any code can go inside the block:
for number in my_favorite_numbers:
    print(number * 10)
````

Notice that just like `A - 10` earlier didn't change the value of `A`, this loop displayed new values but didn't change `my_favorite_numbers` itself.

#### What You Can Do With Booleans: `if` and `else` Statements

Boolean values are most useful for writing **flow control**, or **`if/else` statements.** The general form of a flow control statement in Python is a condition followed by an action clause:

`if this condition is met:`<br />
&nbsp; &nbsp; &nbsp; &nbsp;`perform this action`

The condition is given followed by a colon (`:`). The action clause then follows on the next line, indented into a code block — the same indentation pattern we just used for `for` loops.

* If the condition is fulfilled (evaluates to `True`), the action clause in the block of code is executed.
* If the condition is not fulfilled (evaluates to `False`), the action clause in the block of code is skipped over.

In the cell below, we first assign the value `True` to a variable `sunny_skies`, then we follow it with code that checks if `sunny_skies` is true and takes an action (in this case, a `print()` statement).

````python
# Set a variable to True:
sunny_skies = True

# Check if sunny_skies is True and then take an action if so:
if sunny_skies:
    print("Let's go outside!")
````

**What do you think would happen if you set `sunny_skies` to `False` instead?**

If we want to make sure that something else happens if `sunny_skies` is `False`, we can add an `else` block to our code, which will do something if the `if` condition is not met.

````python
# Set a variable to False:
sunny_skies = False

# Check if sunny_skies is True and then take an action if so:
if sunny_skies:
    print("Let's go outside!")
# Use an else block to take a different action in all other cases:
else:
    print("Let's stay inside today.")
````

An `else` statement **does not require a condition** to evaluate to **True** or **False**. It simply executes when none of the previous conditions are met.

Remember, comparison statements like `10 > 9` resolve to a boolean value. That means we can use them directly in `if` statements without having to directly assign `True` or `False` to a variable first.

````python
if 10 > 9:
    print("Ten is greater than nine.")
else:
    print("Ten is not greater than nine.")
````

Oftentimes, we first assign values to variables before comparing them:

````python
student_grade = 90
passing_grade = 80

if student_grade >= passing_grade:
    print("Congratulations! You passed.")
else:
    print("You shall not pass!")
````
